```python
#Aggregating the value related to an item and grouping by area by calculating its mean.
average_yield = df[df['Item'] == 'Maize'].groupby('Area')['Value'].mean()
print(average_yield)
```

```python
#Derived index and combined area year and items and add to the data set.
df['D_Index'] = df['Area'] + "-" + df['Year'].astype(str) + "-" + df['Item']
df.head()
```

```python
#Calculate Z-score (mean, SD) also.
mean_value = df['Value'].mean()
std_value = df['Value'].std()
df['Z-Score'] = (df['Value'] - mean_value) / std_value

# Define a new threshold for direct Z-Score values.
lower_threshold = -5.0
upper_threshold = 5.0

# Flag rows as outliers based on the direct Z-Score thresholds
df['Outlier'] = (df['Z-Score'] < lower_threshold) | (df['Z-Score'] > upper_threshold)

sorted_data = df.sort_values(by='Z-Score', ascending=True)
outliers = sorted_data[sorted_data['Outlier']]
outliers.head(3)
```
