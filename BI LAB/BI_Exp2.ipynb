```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sales_data.csv', encoding="ISO-8859-1")
print("Done reading")
```

```python
# Convert ORDERDATE to date time format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
# Check for missing values
print("Missing values: ")
print(df.isnull().sum())
```

```python
df.describe
```

```python
print(df.info())
```

```python
# Sales distribution
plt.figure(figsize=(10, 3))
sns.histplot(df['SALES'], kde=True, bins=10, color='blue')
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Sales by year
plt.figure(figsize=(10, 3))
sales_by_year = df.groupby('YEAR_ID')['SALES'].sum()
sales_by_year.plot(kind='bar', color='green')
plt.title("Sales by Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(10, 3))
plt.scatter(df['QUANTITYORDERED'], df['SALES'], alpha=0.5)
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Quantity Ordered vs Sales')
plt.show()
```

```python
top_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False)
print("\nTop Customers by Sales:")
print(top_customers)
```
